```python
m = folium.Map(location=[37.77, -122.44], zoom_start=12)
for index, row in dev_13.iloc[-1250:].iterrows():
    folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=1,
                        popup=str(row['Address']),
                        fill=True,
                        color="#014ccc", # divvy color
                       ).add_to(m)
dev_13_arr = dev_13[['Latitude', 'Longitude']].iloc[-1250:].as_matrix()
m.add_child(plugins.HeatMap(dev_13_arr, radius=15))
m
```
